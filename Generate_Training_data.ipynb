```python
df=pd.read_csv('/content/drive/MyDrive/Jefferson/GNN/AMPS1.csv')
```

```python
df.head()
```

```python
p=df['AirVelocity']*10
```

```python
df
```

```python
AirTemp_Inside = []
for each in df['AirTemp_Inside']:
    AirTemp_Inside.append(str(each).split('.')[0])

AirHumidity_Inside = []
for each in df['AirHumidity_Inside']:
    AirHumidity_Inside.append(str(each).split('.')[0])

AirTemp_Out = []
for each in df['AirTemp_Out']:
    AirTemp_Out.append(str(each).split('.')[0])

AirHumidity_Out = []
for each in df['AirHumidity_Out']:
    AirHumidity_Out.append(str(each).split('.')[0])

Soil_Humidity = []
for each in df['Soil_Humidity']:
    Soil_Humidity.append(str(each).split('.')[0])

Co2_Inside = []
for each in df['Co2_Inside']:
    Co2_Inside.append(str(each).split('.')[0])

Co2_Outside = []
for each in df['Co2_Outside']:
    Co2_Outside.append(str(each).split('.')[0])

VOC_Inside = []
for each in df['VOC_Inside']:
    VOC_Inside.append(str(each).split('.')[0])

VOC_Outside = []
for each in df['VOC_Outside']:
    VOC_Outside.append(str(each).split('.')[0])

Light = []
for each in df['Light']:
    Light.append(str(each).split('.')[0])

AirVelocity = []
for each in p:
    AirVelocity.append(str(each).split('.')[0])

```

```python
poly_df = pd.DataFrame({'AirTemp_Inside':AirTemp_Inside,
                        'AirHumidity_Inside':AirHumidity_Inside,
                        'AirTemp_Out':AirTemp_Out,
                        'AirHumidity_Out':AirHumidity_Out,
                        'Soil_Humidity':Soil_Humidity,
                        'Co2_Outside':Co2_Outside,
                        'VOC_Inside':VOC_Inside,
                        'VOC_Outside':VOC_Outside,
                        'Light':Light,
                        'AirVelocity':AirVelocity,
                        'Co2_Inside':Co2_Inside,
                        })
```

```python
poly_df
```

```python
poly_df['Values'] = poly_df[['AirTemp_Inside', 'AirHumidity_Inside', 'AirTemp_Out', 'AirHumidity_Out', 'Soil_Humidity', 'Co2_Outside', 'VOC_Inside', 'VOC_Outside', 'Light', 'AirVelocity']].apply(lambda x: ', '.join(x[x.notnull()]), axis = 1)
```

```python
new_df=poly_df[['Values','Co2_Inside']]
```
